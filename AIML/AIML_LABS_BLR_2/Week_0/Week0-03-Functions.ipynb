```python
# some more functions
def isPrime(n):
    if n in [2, 3]:
        return True
    if n % 2 == 0:
        return False
    r = 5
    while r * r <= n:
        if n % r == 0:
            return False
        r += 2
    return True
```

```python
print(isPrime(78), isPrime(79))
```

```python
def vowelCount(s):
    """Counts the number of vowels in s and returns the same
    Ignores case of letters"""
    VOWELS = "aeiouAEIOU"
    vc = 0
    for ch in s:
        if ch in VOWELS:
            vc += 1
    return vc
```

```python
print(vowelCount("The quick brown fox jumps over the lazy dog"))
```

#### Recursion
Recursion is also pretty simple.

```python
def isPowerOf2(n):
    """Checks if a number is a power of 2. 
    That is, verifies if an integer p exists such that n == 2 ** p
    Remember 1 is 2 ** 0    """
    if n == 1:
        return True
    if n % 2 == 1:
        return False
    return isPowerOf2(n // 2)
```
