#** Tensor Operations**

###Addition of two tensor

```python
x = torch.rand(3,3)
y = torch.rand(3,3)
z = x+y
print(z)
```

###Addition of two tensor (alternative)

```python
z = x.add(y)
print(z)
```

###Inplace addition of two tensor

```python
print(x)

x.add_(y)

print(x)
```

###Element wise multiplication

```python
print(x*y)
```

###Matrix multiplication

```python
z = torch.matmul(x,y[:,0])
print(z)
z.size()
```

###Matrix multiplication (another example)
####The last two dimensions of the Tensors should be matrix multiplicable and the rest of the existing dimension should match

```python
x = torch.rand(5,4,6,2,2)
y = torch.rand(6,2,5)
z= torch.matmul(x,y)
print(z.size())
```

###Reshaping tensors

```python
x = torch.rand(5,6)
print("Size of x :\n",x.size())
print("x:\n",x)
y = x.view(30)
print("Size of y :\n",y.size())
print("y:\n",y)
z = x.view(-1,10)
print("Size of z :\n",z.size())
print("z:\n",z)
```

###Generating Numbers in Sequence

```python
x = torch.arange(9)
x = x.view(3,3)
print(x)
```

###Concatenating two tensors at a given dimension

```python
print(torch.cat((x,x),dim=0))
print(torch.cat((x,x),dim=1))
```

###Reducing the redundant dimension by squeeze operation

```python
x1= torch.rand(1,5)
print("size of x1 before squeezing:\n",x1.size())
x1 = x1.squeeze()
print("size of x1 before squeezing:\n", x1.size())
```

```python
x2 = torch.rand(3,4,1,5)
print("size of x2 before squeezing:\n",x2.size())
x2 = x2.squeeze()
print("size of x2 before squeezing:\n", x2.size())
```

###Extending the dimension by unsqueeze operation

```python
x = torch.rand(3,4,5,6)
x.unsqueeze(dim=4).size()
```

###Numpy Bridge
####Converting a numpy array to torch tensor

```python
x = np.array([[1.0,2],[3,4]])
y = torch.from_numpy(x)
print(y)
y.type()
```

####Converting a torch tensor to numpy array

```python
z = y.numpy()
z.dtype
```
